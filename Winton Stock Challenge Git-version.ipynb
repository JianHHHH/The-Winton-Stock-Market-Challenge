## different time-series ML problem(learning task)
    

Why time series does not fit for typical ML problem?
-	An intrinsic characteristic of time series is that observations at different time, (instance/row of the dataset) are statistically depend on past observation. So, they don’t naturally fit into the standard machine learning setting where we assumne to have i.i.d instances datasets.
-	In multivariate data, it is implausible to assume that the different univariate component time series are independent and identically distributed
-	In panel data, it is plausible to assume that the different instances are i.i.d., while time series observations within a give instance may still depend on past observation.


Why to understand what kind of problem do I deal with is ABSOLUTELY important is that?
-	Different tasks requires different types of alogirthims to solve them(e.g. times series classfier or forecasters)
-	If you misdiagnose the learning task associated with the real-world problem you’re trying to solve, your algorithm may not work when you want to deploy it.
-	What’s more, if you misdiagnose the task, you can still obtain performance estimates but they may be unrealiable. Performance estimates tell us how well our model will perform when we deploy it. they allow us to make informed choices about which model to deploy. But our estimates are only reliable, if we properly take into accounts the statistical dependence of the data we use to obtain our estimates.

Before machine learning can be used, time series forecasting problems must be re-framed as supervised learning problems. From a sequence to pairs of input and output sequences.

slide woindow can be used for
    
    -transform a single time-series to be x1 y ML problem, where y is the next time step value of x1
    -transform a mulitvariate times series to be x1 x21 x22 y ML problem,where x1 is the previous step of y; and x21,x22 is the previous value and current value of Exogenous variables.
    -transform a univariate time-series multi-step problem to be x1 y1 y2 ML problem, where y1 and y2 are the next one and two time step value of x1.

sliding window method is use of prior time steps of varialbe(s) to predict the future steps. In statistics and time series analysis, this is called a lag or lag method.

**ML notes:**

Most machine learning models do not directly support the notion of observations over time. Instead, the lag observations must be treated as input features in order to make predictions.
This is a benefit of machine learning algorithms for time series forecasting. Specifically, that they are able to support large numbers of input features. These could be lag observations for one or multiple input time series.
Other general benefits of machine learning algorithms for time series forecasting over classical methods include:
    
    •	Ability to support noisy features and noise in the relationships between variables.
    •	Ability to handle irrelevant features.
    •	Ability to support complex relationships between variables.

A challenge with this dataset is the need to make multi-step forecasts. There are two main approaches that machine learning methods can be used to make multi-step forecasts; they are:

    •	Direct. A separate model is developed to forecast each forecast lead time.
    •	Recursive. A single model is developed to make one-step forecasts, and the model is used recursively where prior forecasts are used as input to forecast the subsequent lead time.


Key to the use of machine learning algorithms for time series forecasting is the choice of input data. We can think about three main sources of data that can be used as input and mapped to each forecast lead time for a target variable; they are:
    
    •	Univariate data, e.g. lag observations from the target variable that is being forecasted.
    •	Multivariate data, e.g. lag observations from other variables (weather and targets).
    •	Metadata, e.g. data about the date or time being forecast.

technical knowlege:

Outliers can have many causes, such as:
- Measurement or input error.
- Data corruption.
- True outlier observation (e.g. Michael Jordan in basketball).
There is no precise way to define and identify outliers in general because of the specifics of each dataset. Instead, you, or a domain expert, must interpret the raw observations and decide whether a value is an outlier or not.

detection option:

**1.standard deviation**
If we know that the distribution of values in the sample is **Gaussian or Gaussian-like**, we can use the standard deviation of the sample as a cut-off for identifying outliers. since:

    1 Standard Deviation from the Mean: 68%
    2 Standard Deviations from the Mean: 95%
    3 Standard Deviations from the Mean: 99.7%

3 standard deviations from the mean is a common cut-off in practice. For smaller samples of data, perhaps a value of 2 standard deviations (95%) can be used, and for larger samples, perhaps a value of 4 standard deviations (99.9%) can be used.

**2.Interquartile range**
The IQR can be used to identify outliers by defining limits on the sample values that are a factor k of the IQR below the 25th percentile or above the 75th percentile. The common value for the factor k is the value 1.5. A factor k of 3 or more can be used to identify values that are extreme outliers

**3.Automatic Outlier Detection**
an approach to tackling the problem of outlier detection is one-class classification.One-Class Classification, or OCC for short, involves fitting a model on the “normal” data and predicting whether new data is normal or an outlier/anomaly. This can work well for feature spaces with **low dimensionality (few features)**, although it can become less reliable as the number of features is increased, referred to as the curse of dimensionality.

### 4.6.1 build function to revome ouliters and split X and Y


```python

def outlier_treat_split_X_Y(data):
    """
    remove all the outliers within a dateset use IQR method,
    and split the dataset into X_train and Y_train
        
    """
    
    dataset=data.copy()
    
    #total_features = dataset.columns
    total_features = ['Feature_1','Feature_2','Feature_3','Feature_4','Feature_5','Feature_6','Feature_7','Feature_8','Feature_9',
                  'Feature_10','Feature_11','Feature_12','Feature_13','Feature_14','Feature_15','Feature_16','Feature_17','Feature_18',
                  'Feature_19','Feature_20','Feature_21','Feature_22','Feature_23','Feature_24','Feature_25']

    total_outlier_id=[]

    for feature in total_features:

        q25 = dataset[feature].quantile(0.25)
        q75 = dataset[feature].quantile(0.75)
        iqr = q75 - q25
        # calculate the outlier cutoff, here we use factor 3 because if we use 1.5 we will delet too much instance
        cut_off = iqr * 4
        
        lower, upper = q25 - cut_off, q75 + cut_off

        # identify outliers, since we need to get the index of each outlier instance in order to delete them later on and 
        # id and index are closely related index = id-1; Therefore, we first compare the value to determine if it is a outlier, the
        # collected the outlier's id.

        outliers_id = [i for i,x in dataset[['Id',feature]].values if x < lower or x > upper]
        #print('Identified outliers: %d' % len(outliers_id))

        ###note: there I used extend instead of append, becareful the difference)
        ###total_outlier_index.append(outliers_index)

        total_outlier_id.extend(outliers_id)
        ###print(len(total_outlier_index))

    #there will be duplicated outlier id, since we cannot delete same instance twice, let's get the unique id in the set
    total_outlier_unique_id=np.unique(total_outlier_id)

    #let's get the length to compare if it match with what left
    #print(len(total_outlier_unique_id))

    #as we explained before, in order to delete the outlier instance, we need to get their row index, 
    #since row index = id-1; so we get
    outlier_index=total_outlier_unique_id-1

    dataset.drop(outlier_index, axis=0,inplace=True)
    
    X_train=dataset.loc[:, 'Feature_1':'Ret_120']
    Y_train=dataset.loc[:, 'Ret_121':'Ret_PlusTwo']
    Weights_train=dataset.loc[:,'Weight_Intraday': 'Weight_Daily']

    return X_train, Y_train, Weights_train


    

```

### 4.6.2 bulild function to automatically remolve feature base on missing percentage

```python
# remove columns with few unique values
def Xclean_transform(X, thres_perc=18, verbose=True):
    
    """
    clean/delete features base on missing value percentage of the attributes
        
        Parameters
        ----------
         X : pandas DataFrame, shape [n_samples, n_columns]
            DataFrame contain X_train
thres_perc : integer
             Threshold of missing percentage in order to delete missing attribute
            
        Returns
        -------
        result : clean X_train which delete the missing feature above the threshold

        """
    
    missing_perc=[]
    for i in range(X.shape[1]):
    # count number of rows with missing values
        n_miss = X.iloc[:,[i]].isnull().sum()
        perc = n_miss / X.shape[0] * 100
        missing_perc.append(perc)
    
    #if verbose:
        #print('missing percentage: %s' % missing_perc)
    
    # select columns to delete
    to_del=[ i for i, v in enumerate(missing_perc) if v.values>=thres_perc]
    
    if verbose:
        print('Deleting: %s' % X.columns[to_del])
    
    if len(to_del) is 0:
        return X
    # select all but the columns that are being removed
    
    result = X.drop(columns=X.columns[to_del])
    
    return result
```

### 4.6.3 build customer tranformer for data engineering option1, fix statistic features


```python

from sklearn.base import BaseEstimator, TransformerMixin

#build up transformer for engineering statistic summary feature of time-series value
class add_statistic_features3(BaseEstimator, TransformerMixin):
    
    """
    Add statistic features to the dataframe.
    Add the min, max, mean, and std of each instance of the time-series part to it
    
    """
    
    def __init__(self, **kwargs):
        """initiate_statistic_features to the instance of this class"""
        
        self.mean=kwargs.get('mean') 
        self.min=kwargs.get('mini') 
        self.max=kwargs.get('maxi') 
        self.std=kwargs.get('std') 
        
    def fit(self, X, y=None):

    
        return self

        
    def transform(self, X, y=None):
        
        #make a true copy of the passed dataset
        Xo=X.copy()
        
        if self.mean!=None:
            #calculate the statistic mean of each instance's time-series part, then assign as a new feature
            #to the copied dataset
            Xo[self.mean]=X.filter(regex='Ret_').mean(axis=1)
        
        if self.min!=None:
            Xo[self.min]=X.filter(regex='Ret_').min(axis=1)
        
        if self.max!=None:
            Xo[self.max]=X.filter(regex='Ret_').max(axis=1)
        
        if self.std!=None:
            Xo[self.std]=X.filter(regex='Ret_').std(axis=1)
        
        return Xo
    
    
```

### 5.1.1. b general funtion to  calcualted WMAE with only numpy

```python
###########since skit learn only work with numpy, therefore, I will transfrom the above function in to numpy type

#import modules
from numpy import absolute
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
import numpy as np

# since sckit learn do not provide weghited mean absolute error score, therefore we need to define it by ourself
from sklearn.metrics import make_scorer

#let's define the function to calculated weighted mean absoluted error

def my_WMAE_score_numpy(Y_true, Y_pred, df_weights):
    """ 
    define our customized weighted mean absolute error function
    since it is a measurement of error between true and predict value, the smaller the better, so this is
    a loss func
    
    Parameters
    ----------
    Y_true : array or dataframe
            contain true value of target varible. 
    
    Y_predict : array or dataframe
            contain predictive value of target varible. 
    
    Weights : array or dataframe
            contain Weight_Intraday and Weight_Daily for intraday and daily returns
    
    
    """
    weights=np.array(df_weights)
    
    Y_pred=np.array(Y_pred)
   
    Y_true=np.array(Y_true)
    
#   let's get the number of instance n in the dataset which we predicted
    n=Y_true.shape[0]
#   let's get the number of time-steps value m within each instance 
    m=Y_true.shape[1]
    
    MAE=0
#     #let's set up loops to calculate the weighted mean absolute error
    for i in range(n):
        error=0
        #since with in each instace we have two different weights so we need to 
        for j in range(m):
            
            if j<m-2:
                error=weights[i,0]*abs(Y_true[i,j]-Y_pred[i,j])
               
            else:
                error=weights[i,1]*abs(Y_true[i,j]-Y_pred[i,j])
           
          
            MAE=MAE+error
    
    WMAE=(MAE/n)

    
    return WMAE


#lets make it as a score that can be pass to our cross validation score function later on
```

### 5.1.2.b Define our custimized mean_cross_validation_WMAE function to evaluation each model's performance

```python


def mean_cross_validation_WMAE(ML_model, X_train, Y_train, Weights, k):

    """ 
    define our customized mean_cross_validation_weighted_mean_absolute error function
   
    
    Parameters
    ----------
    ML_model : machine learning model in sckit_learn
               contain true value of target varible. 
    
    X_train: array or dataframe
            contrain value for all the features of the training set
    
    Y_train : array or dataframe
            contain predictive value of target varible. 
    
    Weights : array or dataframe
            contain Weight_Intraday and Weight_Daily for intraday and daily returns.
    
           k: int
           define the number of kold the dataset will be split into for cross-validation.
    """
    
    #define how many folds we will split our data into
    kf = KFold(n_splits=k)
    
    #initialized the dictionary to contain the train and validation index for each iteraion
    t_i={}
    v_i={}
    index=0
    
    #assgine indext to dictionary key values pair
    for train_index, valid_index in kf.split(Y_train):
        t_i[index]=train_index
        v_i[index]=valid_index
        index=index+1
      
    #defint t eamply list to contain each WAME for each iteration
    WMAE_score_list=[]

    for i in range(k):

        ML_model.fit(X_train[t_i[i]], Y_train.iloc[t_i[i]])

        validset_Y_pred=ML_model.predict(X_train[v_i[i]])

        WMAE_score=my_WMAE_score_numpy(Y_train.iloc[v_i[i]], validset_Y_pred, Weights_train.iloc[v_i[i]])

        WMAE_score_list.append(WMAE_score)

    return WMAE_score_list, mean(WMAE_score_list)
```

**What Is Model Selection?**

Model selection is the process of selecting one final machine learning model from among a collection of candidate machine learning models for a training dataset.

Model selection is a process that can be applied both across different types of models (e.g. logistic regression, SVM, KNN, etc.) and across models of the same type configured with different model hyperparameters (e.g. different kernels in an SVM).

Firstly, we need to get over the idea of a “best” model.

All models have some predictive error, given the statistical noise in the data, the incompleteness of the data sample, and the limitations of each different model type. Therefore, the notion of a perfect or best model is not useful. Instead, we must seek a model that is “good enough.”


Therefore, a “good enough” model may refer to many things and is specific to your project, such as:

- A model that meets the requirements and constraints of project stakeholders.
- A model that is sufficiently skillful given the time and resources available.
- A model that is skillful as compared to naive models.
- A model that is skillful relative to other tested models.
- A model that is skillful relative to the state-of-the-art.

Some algorithms require specialized data preparation in order to best expose the structure of the problem to the learning algorithm. Therefore, **we must go one step further and consider model selection as the process of selecting among model development pipelines**.


Instead, there are two main classes of techniques to approximate the ideal case of model selection; they are:

**Probabilistic Measures**: Choose a model via in-sample error and complexity.
Probabilistic measures involve analytically scoring a candidate model using both its performance on the training dataset and the complexity of the model. Four commonly used probabilistic model selection measures include:

- Akaike Information Criterion (AIC).
- Bayesian Information Criterion (BIC).
- Minimum Description Length (MDL).
- Structural Risk Minimization (SRM).

**Resampling Methods**: Choose a model via estimated out-of-sample error.Resampling methods seek to estimate the performance of a model (or more precisely, the model development process) on out-of-sample data.Three common resampling model selection methods include:

- Random train/test splits.
- Cross-Validation (k-fold, LOOCV, etc.).
- Bootstrap.

# self-learning note: esemble method

Ensemble learning is a general meta approach to machine learning that seeks better predictive performance by combining the predictions from multiple models.

the 3 main class of esemble methods:

- Bagging involves fitting many decision trees on different samples of the same dataset and averaging the predictions. the 3 implement step of bagging:
    - Bootstrap samples of the training dataset.
    - Unpruned decision trees fit on each sample.
    - Simple voting or averaging of predictions.

note1: Bagging does not always offer an improvement. For low-variance models that already perform well, bagging can result in a decrease in model performance.

note2: based on note once, there are other ML agorhim can serve as the base model of bagging such as KNN with lower k number also can be used as the base model.

note3:similary esemble type are: Pasting Ensemble,Random Subspaces Ensemble, and Random Patches Ensemble



- Stacking involves fitting many different models types on the same data and using another model to learn how to best combine the predictions. the 3 key elements of this method:
    - Unchanged training dataset.
    - Different machine learning algorithms for each ensemble member.
    - Machine learning model to learn how to best combine predictions.
    
    
- Boosting involves adding ensemble members sequentially that correct the predictions made by prior models and outputs a weighted average of the predictions. the 3 elements of boosting:
    - Bias training data toward those examples that are hard to predict.
    - Iteratively add ensemble members to correct predictions of prior models.
    - Combine predictions using a weighted average of models.

## 6.1 define customized WMAE score for gridsearchcv function.
due to we have a dataset it has two different weight within a samplle, this make the grid search cannot directly apply to our dataset; there are two way to fix it: 
- define our own grid search
- make a customized score function

since we want to use the scikt learn gridsearchcv's inherenty methtod and attributes, **we choose plan B build our customized WMAE score function to pass to gridsearchcv**; we can pass our sample weight argument by two ways: 

option1: set sample_weight parameter in the customized score function, then use make_scorer function's keywards argument option to set sample_wights.

## 6.6 stack ensemble method.
Stacking involves using a machine learning model to learn how to best combine the predictions from contributing ensemble members:

stacking actually is the extension of voting ensemble methods where you try to combine diverse ML algorhim together by taking the average results of each model, stacking try not to average the result of each model, but build a meta ML model use to learn how to assign associate weights to each diverse model, to get a better combine result.

Terminology: the meta_model which build on top of the other model is level 1 model, the ML model use to train the meta-model called the base_model.

The meta-model is often simple, providing a smooth interpretation of the predictions made by the base models，The use of a simple linear model as the meta-model often gives stacking the colloquial name “blending.”

```python
from sklearn.ensemble import StackingRegressor
from sklearn.model_selection import cross_validate


estimators = [
        ('Elastic_Net', ElasticNet(alpha=0.3, fit_intercept=False)),
        #('ExTree', ExtraTreesRegressor(min_samples_leaf=4, n_estimators=70)),
        ('KNN', KNeighborsRegressor(leaf_size=20, n_neighbors=13, p=1)),
        #('gradient_boost_h', HistGradientBoostingRegressor(max_bins=200,max_iter=120))
        #('gradient_boost_h', MultiOutputRegressor(estimator=HistGradientBoostingRegressor(max_bins=200,max_iter=120)))
    ]

#note: when try to blend/stacking all the estimator together use stackingregressor, it only support single target varialbe
# therefore we must apply multioutputregressor to the stacking function to fit multiple output problem; to not double cross
# we nee to delete the multioutput regressor infrom the base estimator of the stacking model

Blend_model = StackingRegressor(estimators=estimators, final_estimator=(LinearRegression()), cv=3, n_jobs=-1)

```
